```
with open('data.pickle', 'rb') as f:
   data_list = pickle.load(f)
```

```
nlp = Okt()
nouns = nlp.nouns(data_list[0])
nouns2 = " ".join(nouns)
```

```
# 워드 클라우드를 설정합니다.
wordcloud = WordCloud(max_font_size=200, font_path='NanumSquareL.ttf',
                      stopwords=spwords,
                     background_color='#FFFFFF',
                     width=1200,height=800).generate(nouns2) # 명사만

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)#TODO  
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
```
